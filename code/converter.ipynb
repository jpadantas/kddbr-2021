```r
library(tidyverse)
library(glue)
n <- "17" # it depends on which samples you are looking for
```

```r
list.files(glue("../outputs/outputs_{n}"))
```

```r
path <- glue('../outputs/outputs_{n}')
```

```r
size_data <- read_csv("../outputs/TSP2Ksizes.csv", col_types = cols(.default = col_integer()))
```

```r
data <- read_csv(glue("../outputs/outputs_{n}/output_final.csv"), col_types = cols(.default = col_integer()))
```

```r
data <- data %>% left_join(size_data, by = "graph")
```

```r
data
```

```r
data <- data %>% pivot_longer(-c(graph, size), names_to = "edge", values_to = "value")
```

```r
data
```

```r
data <- data %>% 
    mutate(edge = str_sub(edge, 3)) %>%
    mutate(from = as.numeric(str_replace(edge, "[^,]+,", ""))) %>% 
    mutate(to = as.numeric(str_replace(edge, ",.*", ""))) %>% 
    select(-edge)
```

```r
data
```

```r
data <- data %>% filter(from < size & to < size & from != to)
```

```r
data
```

```r
data <- data %>% mutate(Id = paste0(graph, ":", from, "-", to))
```

```r
data
```

```r
data <- data %>% select(Id, Predicted = value)
```

```r
print(data)
```

```r
write.csv(data,glue("../solutions/submission_{n}.csv"), row.names = FALSE)
```
